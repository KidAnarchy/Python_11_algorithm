## Coffee Shop

## Optimization model to solve problem.

### Decision Variables:
* Americano = A
* Latte = B
* Cappucino = C
* Mocca = D

### Constrains :
* Coffee: 25A + 25B + 25C + 25D = 2000 cc.
* Hot Water: 75A = 2000 cc.
* Stream Milk: 50B + 25C + 25D = 1500 cc.
* Milk foam: 25B + 50C +25D = 1500 cc.
* Hot Chocolate: 25D = 500 cc.


* A ≥ 0
* B ≥ 0
* C ≥ 0
* D ≥ 0

### Objective Function :
* MAX (25A+30B+30C+30D)

```python
%load_ext iminizinc
```

```python
%%minizinc -m bind 

% Decision Variables
var 0..100: A;  % A is Cup of Americano
var 0..100: B;  % B is Cup of Latte
var 0..100: C;  % C is Cup of Cappucino
var 0..100: D;  % D is Cup of Mocca    
var int:z; % z is income of selling Coffee

% Constraint
% Fresh_Coffee
constraint 25*A + 25*B + 25*C + 25*D <= 2000;
% Hot Water
constraint 75*A <= 2000;
% Stream Milk
constraint 50*B + 25*C + 25*D <= 1500;
% Milk foam
constraint 25*B + 50*C +25*D <= 1500;
% Hot Chocolate
constraint 25*D <= 500;

% Coffee
constraint A>=0;
constraint B>=0;
constraint C>=0;
constraint D>=0;


% Solve problem
solve maximize (25*A + 30*B + 30*C + 30*D);
z = (25*A + 30*B + 30*C + 30*D);
```

```python
print('\n-----------------SUMMARY--------------------')
print("\nAmericano =",A,'Cups')
print("Latte =",B,"Cups")
print("Cappucino =",C,'Cups')
print("Mocha =",D,"Cups")
print("max total income =",z,"Bath")

print('\n--------------------------------------------')
```
