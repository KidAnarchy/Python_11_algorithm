## Optimization model to solve problem.

### Decision Variables:
* Vanilla = X
* Stawberry = Y

### Constrains :
* FreshMilk: 0.5X+0.2Y = 10
* Doll : X + Y  30
* X  0
* Y  0

### Objective Function :
* MAX (2X+3Y)

```python
%load_ext iminizinc
```

```python
%%minizinc -m bind   
% Decision Variables
var 0..100: X;  % X is Box of Vanilla
var 0..100: Y;  % Y is Box of Stawbeery
var int:z; % z is income of selling Icecream

% Constraint
% FreshMilk
constraint 0.5*X + 0.2*Y <= 10;
% Dolls
constraint X + Y <= 30 ;
% IceCream
constraint X>=0;
constraint Y>=0;

% Solve problem
solve maximize (2*X + 3*Y);
z = (2*X + 3*Y);
```

```python
print('\n-----------------SUMMARY--------------------')
print("\nVanilla =",X,'Boxes')
print("Stawberry =",Y,"Boxes")
print("max total income =",z,"USD")
print('We can give dolls to customer:',X + Y,'dolls')
print('\n--------------------------------------------')
```
