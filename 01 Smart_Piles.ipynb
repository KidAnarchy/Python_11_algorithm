```python
# 003_Smart_Piles

print("\n==Take stick from pile Human VS The Smartest Computer==")
N = int(input('\nHow many sticks (N) in the pile:'))   # N = amount of stick in the pile
print('There are', N, 'sticks in the pile.')
name = str(input('What is your name :'))             # Player 1 Somchai or Your name:
print('\nwho take the last stick lose!')
print('--------------------------------------------------')

# Player2: Python the smartest computer:
for Y in range (N):     # Y = take a stick from the pile by Computer.
    print('\nI am the smartest computer, take :')
    
# Constraint: for take a stick to make player1 get a spacial number and always loss.
    if N > 1 and N % 3 == 0:
            N = N - 2
            print('2')
    elif N > 1 and N % 3 != 0:
            N = N - 1
            print('1')    
    else:
            print('The last stick',name, 'win (I, smart computer, am sad T_T)')
            break               
    print('There are', N, 'sticks in the pile.')  
    
# Player1:   
    for X in range (N):   # X = take a stick from the pile by Player1. 
        print('\n')
        print(name ,',how many sticks you will take (1 or 2):')
        X = int(input())
# Constraint:    
        if X >=3:
            print('No you cannot take more than 2 sticks!')
            continue
        elif 1<= X <3:
            if X == 2:
                N = N - X  
                if N > 0:
                    break
                else:
                    print('There are not enough sticks to take.')
                    N = X + N
                    if N == 1:
                        print('Game Over',',',name,',','You have to take the last stick, I am the smartest computer, win!!!')
                        N = N - 1
            elif X == 1:
                N = N - X
                if N > 0:
                    break
                else:
                    print('Game Over',',',name,',','You take the last stick, I am the smartest computer, win!!!')
        else:
            print('No you cannot take less than 1 stick!')
            continue
    print('There are', N, 'sticks in the pile.')
    print('----------------------------------------------------')   
                          
    if N == 0:    # If N = 0 is End Game ..
        break 
```
